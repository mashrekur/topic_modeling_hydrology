```python
import numpy as np
import pickle as pkl
import matplotlib.pyplot as plt
from gensim.models import LdaModel
import pandas as pd
from tqdm import tqdm
```

```python
nTopics = 45
```

```python
# Load model
lda_model = LdaModel.load(f'trained_models/trained_lda_model_new_{nTopics}')
```

```python
# Load raw corpus
with open('data/cleaned_corpus_broad.pkl', 'rb') as f:
    corpus = pkl.load(f)
```

```python
len(corpus)
```

```python
transformed_corpus = lda_model[corpus]
```

```python
#transformed corpus [topic_id][document_topic distributions][each document_topic distribution]
transformed_corpus[0][0][0]
```

```python
topic_dist = np.zeros([len(corpus),lda_model.num_topics])
topic_dist.shape
```

```python
for i in tqdm(range(len(transformed_corpus))):
    for j in transformed_corpus[i][0]:
        topic_dist[i, j[0]] = j[1]
        
```

```python
for i in tqdm(range(len(corpus))):
    for j in transformed_corpus[i][0]:
        topic_dist[i, j[0]] = j[1]
```
